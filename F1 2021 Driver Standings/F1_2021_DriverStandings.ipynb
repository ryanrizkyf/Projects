## Import Library

```python
import pandas as pd
import bar_chart_race as bcr
```

## Import Dataset

```python
df = pd.read_excel('F1_2021_DriverStandings.xlsx')
```

```python
df.head()
```

```python
df.info()
```

## Pre-Processing Data

```python
columns = ['Valtteri Bottas',
          'Lando Norris',
          'Sergio Perez',
          'Charles Leclerc',
          'Daniel Ricciardo',
          'Carlos Sainz',
          'Yuki Tsunoda',
          'Lance Stroll',
          'Kimi Raikkonen',
          'Antonio Giovinazzi',
          'Esteban Ocon',
          'George Russell',
          'Sebastian Vettel',
          'Mick Schumacher',
          'Pierre Gasly',
          'Nicholas Latifi',
          'Fernando Alonso',
          'Robert Kubica',
          'Nikita Mazepin']

for col in columns:
    df[col] = df[col].astype(float)

df['Date'] = df['Date'].astype(str)

df = df.set_index('Date')
```

## Bar Chart Race Animation

```python
bcr.bar_chart_race(df = df,
                   n_bars = 21,
                   sort = 'desc',
                   title = 'Formula 1 2021 Driver Standings',
                   filename = 'formula_one_2021_driver_standings.mp4', 
                   bar_label_size=7, 
                   tick_label_size=7, 
                   fixed_order=False, 
                   fixed_max=True, 
                   steps_per_period=10, 
                   interpolate_period=False, 
                   label_bars=True, 
                   bar_size=.95, 
                   period_length=750, 
                   dpi=1000)
```
